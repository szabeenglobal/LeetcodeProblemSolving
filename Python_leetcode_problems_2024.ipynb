## Ismorphic String
Given two strings s and t, determine if they are isomorphic.

Two strings s and t are isomorphic if the characters in s can be replaced to get t.

All occurrences of a character must be replaced with another character while preserving the order of characters. No two characters may map to the same character, but a character may map to itself.

````python
def isIsomorphic(s, t):
    s_map = {}
    t_map = {}
    # Iterate over both strings simultaneously
    for char_s, char_t in zip(s, t):
    # Check if there's an existing mapping from s to t
        if char_s in s_map:
    # If the mapping doesn't match the current character in t, return False
            if s_map[char_s] != char_t:
                return False
        else:
            if char_t in t_map:
                return False
     # Establish the new mapping
            s_map[char_s] = char_t
            t_map[char_t] = char_s
    print(s_map)
    print(t_map)
    return True

isIsomorphic('add', 'egg')
````

Given two strings s and t, return true if t is an 
anagram of s, and false otherwise.
Anargam
a word, phrase, or name formed by rearranging the letters of another, such as spar, formed from rasp.

### Explanation
Let me break down the line of code s_count[char] = s_count.get(char, 0) + 1 in detail for you.

Purpose:
This line of code is used to count the occurrences of each character in the string s. The dictionary s_count stores how many times each character appears in the string.

Breakdown:
s_count.get(char, 0):

This part of the code retrieves the current count of the character char from the dictionary s_count.
If the character char is not yet in the dictionary, s_count.get(char, 0) will return the default value 0 (since the second argument to get() is 0).
If the character is already in the dictionary, it returns its current count.
+ 1:

After getting the current count of char (or 0 if it hasn't been counted yet), you add 1 to the count to indicate that you've encountered this character again.

````python
def isAnagram(s, t):
    s_count = {}
    t_count = {}
    if len(s) != len(t):
        return False
    for char in s:
        s_count[char] = s_count.get(char, 0) +1
    for char in t:
        t_count[char] = t_count.get(char, 0) +1
    return s_count == t_count

s = "anogram"
t = "nagaram"
print(isAnagram(s, t))
````

Given an input string s, reverse the order of the words.
A word is defined as a sequence of non-space characters. The words in s will be separated by at least one space.

Return a string of the words in reverse order concatenated by a single space.

Note that s may contain leading or trailing spaces or multiple spaces between two words. The returned string should only have a single space separating the words. Do not include any extra spaces.

````python
s = "London is a beautiful city"
def rev_words(s):
    w_list = s.split()
    n = len(w_list)
    start = 0
    end = n-1
    while start < end:
        w_list[start], w_list[end] = w_list[end], w_list[start]
        start+=1
        end-=1

    return ' '.join(w_list)
rev_words(s)
````

    How the Function Works Step-by-Step:<br>
    </p>
**Base Case:**<br>

The function continues merging as long as a < b. When a == b, it indicates that all elements in the range [a:b] have been filled, and the function stops. <br>
<b>Select the Larger Element:</b><br>

The function is building the merged section from back to front. This approach allows for placing elements without overwriting any unprocessed elements in A.<br>
The function checks (j <= 0) first to see if there are any elements left in R. If j == 0, it means all elements from R have been merged, so it proceeds to take elements from L.<br>
If there are elements in both L and R, the function compares the last element of each (L[i-1] and R[j-1]). The larger element is placed at A[b-1].<br>
<b>Place the Element in A and Update Indices:</b>

If L[i-1] > R[j-1], it places L[i-1] at the end of the current segment in A and decrements i to move to the next element in L. </br>
Otherwise, it places R[j-1] at the end of the current segment in A and decrements j to move to the next element in R.
After placing an element, it calls itself recursively with b - 1 to fill the next position from the end.

<b>Recursive Call:</b>

The function calls itself again with updated parameters to process the next element until all elements from L and R are merged into A.
###  Function merge_sort:


The merge_sort function sorts the array A in the range [a:b].
If b is not provided, it defaults to the length of A (meaning the entire list).
It uses a divide-and-conquer approach by recursively dividing the array in half, sorting each half, and then merging them. <br>
**Base Case:**
<br>
The function checks if 1 < b - a, which ensures the segment has more than one element. If there is only one element (or none), it returns as this segment is already sorted.<br>
**Recursive Case:**
<br>
c = (a + b + 1) // 2 calculates the middle index of the segment.<br>
🔵 **line 22:** merge_sort(A, a, c) recursively sorts the first half.<br>
🔵 **line 23:** merge_sort(A, c, b) recursively sorts the second half.<br>
L, R = A[a:c], A[c:b] creates two lists, L and R, holding the two halves.<br>
merge(L, R, A, len(L), len(R), a, b) merges the two halves back into the original array A.<br>

````python
def merge(L, R, A, i, j, a, b):
    '''Merge sorted L[:i] and R[:j] into A[a:b]'''
    if a < b:
        if (j <= 0) or (i > 0 and L[i - 1] > R[j - 1]):
            A[b - 1] = L[i - 1]
            i = i - 1
        else:
            A[b - 1] = R[j - 1]
            j = j - 1
        merge(L, R, A, i, j, a, b - 1)
def merge_sort(A, a=0, b=None):
    '''Sorts A[a:b] using merge sort algorithm.'''
    if b is None:
        b = len(A)  # If b is not provided, set it to the length of A

    # Base case: if the segment length is greater than 1
    if 1 < b - a:
        # Find the midpoint of the current segment
        c = (a + b + 1) // 2

        # Recursively sort the left and right halves
        merge_sort(A, a, c)
        merge_sort(A, c, b)

        # Create sublists for the left and right halves
        L, R = A[a:c], A[c:b]

        # Merge the sorted halves back into A
        merge(L, R, A, len(L), len(R), a, b)
````

## Pointers in Different Types of Nodes

In computer science, nodes can have various types and numbers of pointers depending on the data structure they belong to. Here are some common examples:

### 1. **Singly Linked List Node**
   - **Pointers**: 1 pointer (next)
   - **Description**: 
     Each node in a singly linked list contains a single pointer, usually called `next`, which points to the next node in the sequence. The last node’s `next` pointer points to `null` (or `None` in Python) to signify the end of the list.
   - **Diagram**:
     ```
     [Data] -> [Next]
     ```

### 2. **Doubly Linked List Node**
   - **Pointers**: 2 pointers (next, prev)
   - **Description**: 
     A node in a doubly linked list contains two pointers: `next`, which points to the next node, and `prev`, which points to the previous node. This allows traversal in both directions.
   - **Diagram**:
     ```
     [Prev] <-> [Data] <-> [Next]
     ```

### 3. **Binary Tree Node**
   - **Pointers**: 2 pointers (left, right)
   - **Description**: 
     In a binary tree, each node has two pointers: `left` and `right`. These pointers refer to the left and right child nodes, respectively. In a binary search tree (BST), nodes to the left contain values less than the parent, and nodes to the right contain values greater than the parent.
   - **Diagram**:
     ```
       [Data]
       /    \
     [Left] [Right]
     ```

### 4. **N-ary Tree Node**
   - **Pointers**: n pointers (children)
   - **Description**: 
     An N-ary tree node can have up to `n` children. This is common in structures like general trees or in file system representations where nodes may have varying numbers of children.
   - **Diagram** (for N = 3):
     ```
       [Data]
      /  |  \
    [C1][C2][C3]
     ```

### 5. **Graph Node (or Vertex)**
   - **Pointers**: Variable number (depends on connections or edges)
   - **Description**: 
     In an adjacency list representation of a graph, each graph node can have multiple pointers corresponding to each edge connecting it to other nodes. The number of pointers equals the node's degree (number of edges connected).
   - **Diagram**:
     ```
     [Data] --> [Neighbor1] --> [Neighbor2] --> ... --> [NeighborN]
     ```

### 6. **Skip List Node**
   - **Pointers**: Variable (next, multiple levels)
   - **Description**: 
     Skip lists are a probabilistic alternative to balanced trees. Each node may have multiple `next` pointers at different levels, allowing for faster search operations by "skipping" over multiple nodes.
   - **Diagram** (with 3 levels):
     ```
     [L3] -> [L2] -> [L1] -> [Data]
     ```

### 7. **Trie (Prefix Tree) Node**
   - **Pointers**: Fixed number based on the character set (e.g., 26 pointers for English alphabet)
   - **Description**: 
     In a trie, each node represents a character in a word and may have up to a fixed number of pointers, often 26 for the English alphabet. Each pointer represents a possible character transition.
   - **Diagram** (partial):
     ```
       [A]
      / | \
    [B][C][D]
     ```

#### Conclusion
Different data structures use varying numbers of pointers in nodes to maintain relationships between elements efficiently. Understanding these structures helps to choose the right data structure based on traversal and performance requirements.

````python
def sqrt(x): 
    if x == 0 or x == 1:
        return x
    left, right = 0, x//2 +1
    while left<=right :
        mid = (left+right)//2
        square = mid*mid
        if square == x:
            return mid
        elif square < x:
            left = mid + 1
        else:
            right = mid - 1
    return right

sqrt(100)
````

## 50. Pow(x, n) 41/150
 Implement pow(x, n), which calculates x raised to the power n (i.e., xn).
### Explanation of the Code

1. **Handle \( n = 0 \):** If \( n = 0 \), the result is `1` since \( x^0 = 1 \) for any \( x \).

2. **Handle Negative Exponents:** If \( n \) is negative, convert it to positive by taking the reciprocal of \( x \).

3. **Exponentiation by Squaring:**
   - If \( n \) is odd, multiply `result` by `current_product`.
   - Square `current_product` to get the next higher power, and halve \( n \).

4. **Return the Result:** Once \( n \) reaches 0, `result` holds the answer.

#### Here’s Why We Halve \( n \):

##### Binary Representation of Exponent
Exponentiation by squaring leverages the binary representation of \( n \) (e.g., \( 13 = 1101 \) in binary). By iterating over the bits of \( n \), each bit represents whether to include the current power of \( x \) in the result. Halving \( n \) in each step (equivalent to right-shifting in binary) effectively moves us through the bits of \( n \) from the least significant to the most significant.

##### Efficient Power Calculation
Each time we square `current_product`, we double the power it represents. By halving \( n \), we determine if that power is needed:

- **If \( n \) is odd**, we multiply `result` by `current_product` to "include" this power of \( x \).
- **If \( n \) is even**, we skip this multiplication but still square `current_product`.

This reduction strategy minimizes the number of multiplications. Instead of computing \( x^n \) directly with \( n \) multiplications, we achieve it in \( O(\log n) \) steps by squaring and halving \( n \) in each loop iteration.


````python

 def myPow(x,n):
    if not n:
        return 0
    if n == 0:
        return 1
    if n<0:
        # return 1/ self.myPow(x, -n)
        x = 1/x
        n=-n

    result =1
    current_product = x

    while n > 0: 
        if n % 2 == 1:
            result *= current_product
        current_product *= current_product
        
        n //=2
    return result


myPow(2, 13)
````

The key idea is that the leftmost depth provides enough information to help us determine whether a subtree is a perfect binary tree or not. **Case 1:  if left_depth == right_depth:**
If the left and right depths are the same, it means the entire left subtree forms a perfect binary tree up to that depth.
For a perfect binary tree with depth d, the total number of nodes is 2**(d+1) -1

**Case 2: left_depth != right_depth**

If the depths are different, it implies the left subtree has an extra level over the right subtree. This means the right subtree is a perfect binary tree of height right_depth.
Thus, we can calculate the number of nodes in the right subtree as 2**(right_depth+1) -1

## Depth-First Search (DFS)

### Introduction
Depth-First Search (DFS) is an algorithm used for traversing or searching through trees and graphs. 
DFS starts at a source node and explores as far as possible along each branch before backtracking. 
It’s commonly implemented using either recursion (implicit stack) or an explicit stack data structure.

### DFS Algorithm Outline
1. Start at the root (or source) node and push it onto the stack (or call recursively).
2. Pop the node from the stack and process it (e.g., print the node value).
3. Push all unvisited adjacent nodes (or children, in a tree) onto the stack.
4. Repeat steps 2-3 until the stack is empty.

### Code Implementation
### Complexity Analysis
- **Time Complexity**: \(O(V + E)\), where \(V\) is the number of vertices (nodes) and \(E\) is the number of edges.
- **Space Complexity**: \(O(V)\) for the stack (in case of recursive DFS, space is used by recursion stack)


````python
#### Recursive DFS

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def dfs_recursive(node):
    if not node:
        return
    print(node.val)  # Process the node
    dfs_recursive(node.left)  # Recur on the left child
    dfs_recursive(node.right)  # Recur on the right child

#### Iterative DFS (using stack)

def dfs_iterative(root):
    if not root:
        return
    stack = [root]
    while stack:
        node = stack.pop()
        print(node.val)  # Process the node
        if node.right:
            stack.append(node.right)  # Right child pushed first so that left is processed first
        if node.left:
            stack.append(node.left)  # Left child pushed last to be processed first
````

### Introduction
Breadth-First Search (BFS) is an algorithm used to traverse or search trees and graphs. 
BFS explores each level fully before moving to the next level, which makes it optimal for finding the shortest path in an unweighted graph.

### BFS Algorithm USING A QUEUE Outline
1. Start at the root (or source) node, and add it to a queue.
2. Remove the node from the front of the queue, and process it.
3. Add all unvisited adjacent nodes (or children, in a tree) to the end of the queue.
4. Repeat steps 2-3 until the queue is empty.

### Complexity Analysis
- **Time Complexity**: \(O(V + E)\), where \(V\) is the number of vertices (nodes) and \(E\) is the number of edges.
- **Space Complexity**: \(O(V)\) for the queue.


## Summary

- **DFS** is ideal for exploring all paths or when you need to reach all leaf nodes before exploring siblings.
- **BFS** is suitable for finding the shortest path in unweighted graphs and ensures all nodes at the current depth are visited before going deeper.


### Implementation of Breadth first search using a queue

````python
### Code Implementation

from collections import deque

def bfs(root):
    if not root:
        return
    queue = deque([root])
    while queue:
        node = queue.popleft()  # Pop from the front of the queue
        print(node.val)  # Process the node
        if node.left:
            queue.append(node.left)  # Add left child to queue
        if node.right:
            queue.append(node.right)  # Add right child to queue
````

- Pop the top node from the stack and set it as the current node.
- Calculate the difference between current.val and prev if prev is set.
- Update min_diff with this difference if it’s smaller than the current min_diff.
#### Update prev:
- Set prev to the value of the current node to be used for the next node’s comparison.
#### Move to Right Subtree:
- Set current to current.right to continue the in-order traversal on the right subtree.

````python
from collections import deque
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
    

def getMinimumDifference(root):
    
    min_diff = float('inf')
    prev = None
    current = root
    stack = []

    while current or stack:
        while current:
            stack.append(current)
            current = current.left
        current = stack.pop()   # Process the node at the top of the stack
        if prev is not None:
            diff = current.val - prev 
            min_diff = min(min_diff, diff)
        prev = current.val
        current = current.right

    return min_diff
````

- Push Operation: We push val onto stack. For min_stack, we push val if it’s smaller than or equal to the current minimum; otherwise, we push the current minimum again to maintain the minimum history.
- Pop Operation: We pop from both stack and min_stack to keep them synchronized.
- Top Operation: We return the top element of stack.
- GetMin Operation: The current minimum is always at the top of min_stack. <br>
**Complexity** <br>

````python
class MinStack():
    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, val:int) -> None:
        self.stack.append(val)
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)
        else:
            self.min_stack.append(self.min_stack[-1])
    def pop(self)-> None:
        if self.stack:
            self.stack.pop()
            self.min_stack.pop()

    def top(self)  -> int: 
        return self.stack[-1] if self.stack else None
    def getMin(self) -> int:
        return self.min_stack[-1] if self.min_stack else None
### sample usage
minStack = MinStack()
minStack.push(-2)
minStack.push(0)
minStack.push(-3)
print(f" minimum value {minStack.getMin()}") # Returns -3
minStack.pop()
print(minStack.top())    # Returns 0
print(minStack.getMin())
````

# Main Concepts of Graph Theory 

The **main concepts of graph theory** are foundational for many applications in computer science and engineering, including networks, scheduling, and optimization. MIT’s curriculum in graph theory typically emphasizes both the **theoretical foundations** and **practical algorithms** for solving problems. Let’s go over the core concepts:

## 1. Graph Representation
   - **Graph Basics**: A graph is a collection of **nodes (vertices)** connected by **edges**. Graphs can represent various things: road networks, social connections, dependencies, etc.
   - **Types of Graphs**:
      - **Undirected Graph**: Edges have no direction; they simply connect two nodes.
      ```
     A --- B
      \\   /
       C
     ```
      - **Directed Graph (Digraph)**: Edges have direction, going from one node to another.

            ```  
               A → B
               ↑
               C → D
            ```
      - **Weighted Graph**: Each edge has a weight or cost associated with it, useful for representing distances, costs, etc.
            
            ``` A --5-- B
               \      /
               2\    /3
               \  /
                  C 
                   ```
      - **Unweighted Graph**: All edges have equal weight (or no weight at all).
      - **Connected Graph**: There’s a path between every pair of nodes.
               
                ```
               A --- B
               \   /
                  C --- D
                   ```
      - **Disconnected Graph**: Some nodes are unreachable from others.
              
               ```
              A --- B     E --- F
               \   /       \   /
                  C           G
               ```
   - **Graph Representations**:
      - **Adjacency Matrix**: A 2D matrix where each cell `(i, j)` represents an edge between vertices `i` and `j`. This is space-efficient for dense graphs but can be wasteful for sparse graphs.
      - **Adjacency List**: A list of lists or dictionaries where each vertex has a list of adjacent vertices. This is space-efficient for sparse graphs.

## 2. Traversals (Exploring Graphs)
   - **Depth-First Search (DFS)**:
      - Explores as deep as possible along each branch before backtracking.
      - Implemented using recursion or a stack.
      - Used for applications like finding connected components, cycle detection, topological sorting, and solving maze-like problems.

   - **Breadth-First Search (BFS)**:
      - Explores all neighbors at the present "depth" before moving on to nodes at the next depth level.
      - Implemented using a queue.
      - Commonly used for finding the shortest path in an unweighted graph, level-order traversal, and checking bipartiteness.

## 3. Graph Properties and Structures
   - **Paths and Cycles**:
      - **Path**: A sequence of edges connecting a sequence of vertices.
      - **Simple Path**: A path with no repeated vertices.
      - **Cycle**: A path where the first and last vertices are the same.
      - **Acyclic Graph**: A graph with no cycles, important in many applications such as scheduling.

   - **Connectivity**:
      - A graph is **connected** if there’s a path between any two vertices.
      - In directed graphs, **strongly connected** means there’s a path between every pair of vertices in both directions.
      - **Connected Components**: Maximal sets of connected vertices in a graph, useful in clustering.

## 4. Tree Structures
   - **Trees**: A tree is a connected, acyclic undirected graph. It has `n` vertices and `n-1` edges, and there is exactly one path between any two nodes.
   - **Spanning Tree**: A subset of a graph that includes all vertices with minimal edges, forming a tree. Spanning trees are important in network design.
   - **Minimum Spanning Tree (MST)**: A spanning tree with the minimum total edge weight. MST algorithms, like Kruskal’s and Prim’s, are essential for optimizing networks.

## 5. Graph Algorithms
   - **Shortest Path Algorithms**:
      - **Dijkstra’s Algorithm**: Finds the shortest path from a source node to all other nodes in a graph with non-negative edge weights.
      - **Bellman-Ford Algorithm**: Computes shortest paths even with negative edge weights, detecting negative cycles.
      - **Floyd-Warshall Algorithm**: Finds shortest paths between all pairs of nodes, typically used for dense graphs.

   - **Minimum Spanning Tree Algorithms**:
      - **Kruskal’s Algorithm**: Builds the MST by adding the smallest edges that don’t form a cycle.
      - **Prim’s Algorithm**: Starts from a single node and grows the MST by adding the minimum weight edge from the tree to a new vertex.

   - **Topological Sorting**:
      - Ordering vertices in a Directed Acyclic Graph (DAG) so that for every directed edge `(u, v)`, `u` appears before `v`.
      - Used in scheduling tasks, dependency resolution, and course prerequisites.

   - **Graph Coloring**:
      - Assigning colors to vertices such that no two adjacent vertices share the same color.
      - Applications include scheduling, register allocation in compilers, and solving puzzles like Sudoku.

   - **Network Flow**:
      - **Maximum Flow Problem**: Determines the maximum possible flow in a network from a source to a sink.
      - Algorithms like **Ford-Fulkerson** and **Edmonds-Karp** are commonly used for network flow problems, applied in logistics, transportation, and telecommunications.

## 6. Advanced Topics
   - **Eulerian and Hamiltonian Paths**:
      - **Eulerian Path**: A path that visits every edge exactly once. Exists if exactly two vertices have an odd degree.
      - **Hamiltonian Path**: A path that visits every vertex exactly once, much harder to find and often solved using heuristics.
   
   - **Strongly Connected Components (SCCs)**:
      - In a directed graph, SCCs are maximal subgraphs where every vertex is reachable from every other vertex in the subgraph.
      - **Kosaraju’s Algorithm** or **Tarjan’s Algorithm** can efficiently find SCCs, used in analyzing social networks and dependencies.

   - **Graph Matching**:
      - **Matching**: A set of edges where no two edges share a common vertex. Used in pairing problems like assigning jobs to applicants.
      - **Maximum Matching**: The largest possible matching, found using algorithms like the **Hungarian Algorithm**.

## 7. Complexity and Optimization
   - **Big-O Analysis**: Understanding the time and space complexity of graph algorithms is essential, especially for large graphs.
   - **Heuristics**: Approximation techniques for NP-hard graph problems (like Traveling Salesman Problem or finding Hamiltonian cycles).
   - **Approximation Algorithms and Greedy Techniques**: Often used in optimization problems where exact solutions are computationally impractical.
